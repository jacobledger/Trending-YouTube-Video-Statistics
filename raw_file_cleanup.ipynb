```python
# The file allows tags to be seperated by '|' while also quoting them preventing parsing
# Replace "|" between tags with a safe separator inside one quoted field

import re
with open('data/USvideos.csv', 'r', encoding='utf-8') as us:
    content = us.read()

# "tag1"|"tag2"|"tag3"
improperly_quoted = r'"([^"]+)"\|"([^"]+)"'

# Keep replacing until no more "|" between quoted tags exists
while re.search(improperly_quoted, content):
    content = re.sub(improperly_quoted, r'"\1|\2"', content) # →  "tag1|tag2|tag3"

with open('data/USvideos_cleaned.csv', 'w', encoding='utf-8') as us:
    us.write(content)

print("Done!")
```
